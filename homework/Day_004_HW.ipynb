```python
import os
import numpy as np
import pandas as pd
```

```python
# 設定 data_path
dir_data = './data/'
```

```python
f_app = os.path.join(dir_data, 'application_train.csv')
print('Path of read in data: %s' % (f_app))
app_train = pd.read_csv(f_app)
```

```python
app_train.info()
```

### 如果沒有想法，可以先嘗試找出剛剛例子中提到的問題的答案
#### 資料的 row 數以及 column 數

```python
num_rows = len(app_train.index)
num_columns = len(app_train.columns)
print('row數:', num_rows)
print('column數:', num_columns)
```

```python
print('row數:', app_train.shape[0])
print('column數:', app_train.shape[1])
```

#### 列出所有欄位

```python
app_train.columns.values
```

#### 截取部分資料

```python
#擷取特定欄位

columns = ['SK_ID_CURR', 'AMT_INCOME_TOTAL', 'AMT_REQ_CREDIT_BUREAU_HOUR']

app_train[columns].head()
```

```python
app_train.loc[:, columns].head()
```

```python
#從特定欄位條件篩選列(FLAG_OWN_CAR = N)

FLAG_OWN_CAR_N = app_train['FLAG_OWN_CAR'] == 'N'
app_train[FLAG_OWN_CAR_N]
```

```python
#List people whose AMT_INCOME_TOTAL is greater than average

AMT_INCOME_TOTAL = app_train['AMT_INCOME_TOTAL'].values
mean = np.mean(AMT_INCOME_TOTAL)

print('List people whose AMT_INCOME_TOTAL is greater than average (%f):' % (mean))

app_train[AMT_INCOME_TOTAL > mean]
```
