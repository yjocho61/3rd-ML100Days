```python
%matplotlib inline
import matplotlib.pyplot as plt
data = df['縣市'].values
data2 = df['死亡人數'].values.astype(int)
print(data[data2 > 1])
print(data2[data2 > 1])
```

```python
s = dict()

def dict_append(dict, element):
    if element not in dict:
        dict[element] = 0
    return dict[element]

for i in range(len(data)):
    s[data[i]] = dict_append(s, data[i]) + 1 * data2[i]
for key in s:
    print(key, s[key])
```

```python
plt.bar(list(s.keys()), s.values(), color='g')
```
